```python
churn_counts = df['Churn'].value_counts()
plt.bar(churn_counts.index, churn_counts.values, color= 'c')
plt.xlabel("Targte Variable")
plt.ylabel("Count")
```

```python
df['Churn'].value_counts().plot(kind='bar', figsize=(8,6), color='c')
plt.title("Count of Target Variable")
plt.ylabel('Count')
plt.xlabel('Target Variable')
```

Now I would like to turn "YES" and "NO"s into numeric values so I can conduct calculations more easily

```python
df['Churn'] = np.where(df.Churn == 'Yes', 1, 0)
```

```python
df.head(9)
```

```python
x = df['tenure']
y = df['MonthlyCharges']

plt.scatter(x,y)
```

```python
def plot_pairing(data, x_feature, y_feature, title=None):
    plt.figure(figsize=(12,8))
    churn_data = data[data['Churn'] == 1]
    not_churn_data = data[data['Churn'] == 0]

    plt.scatter(churn_data[x_feature], churn_data[y_feature], color= 'red', label='Churned')
    plt.scatter(not_churn_data[x_feature], not_churn_data[y_feature], color='green', label='Not Churned')

    plt.title(title, fontsize=20)
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.legend()
    plt.show()
```

```python
scatter_data = df[['tenure','MonthlyCharges', 'Churn']]
```

```python
plot_pairing(scatter_data, 'tenure', 'MonthlyCharges',  title='Tenure vs. MonthlyCharges')
```

```python
plt.hist(df['tenure'], bins=15, color='c', edgecolor='black')
plt.xlabel('Tenure')
plt.ylabel('Frequency')
plt.title('Distribution of Tenure')
plt.grid()
plt.show()
```

```python
tenure_churn_no = df[df.Churn == 0].tenure
tenure_churn_yes = df[df.Churn == 1].tenure
```

```python
plt.hist([tenure_churn_no, tenure_churn_yes], bins=12, color=['c','r'], edgecolor='black', label=['Not Churned', 'Churned'])
plt.xlabel('Tenure')
plt.ylabel('Frequency')
plt.title('Customer Churn Prediction')
plt.legend()
plt.grid()
plt.show()
```
